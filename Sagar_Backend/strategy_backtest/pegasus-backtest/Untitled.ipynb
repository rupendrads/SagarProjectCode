```python
data = pd.read_csv(r'C:\Users\pegas\OneDrive\Desktop\data\2024\jan\GFDLNFO_BACKADJUSTED_02012024.csv')
data.rename(columns={'Ticker': 'symbol'}, inplace=True)

data['Time'] = pd.to_datetime(data['Time'])
data['Time'] = data['Time'].dt.time
data['datetime'] = pd.to_datetime(data['Date']) + pd.to_timedelta(data['Time'].astype(str))
data.columns = data.columns.str.lower()
data.drop(['date','time'],axis=1, inplace=True)
data['datetime'] = data['datetime'].dt.floor('min')

data.reset_index(inplace=True,drop=True)

datetime_column = data.pop('datetime')
data.insert(0, 'date', datetime_column)


data

data['options_type'] = data['symbol'].str.replace('.NFO', '').str[-2:]
data['options_type'] = data['options_type'].apply(lambda x: 'FUT' if x not in ['PE', 'CE'] else x)
data['strike'] = data['symbol'].str.replace('.NFO', '').str[-7:-2]
data['strike'] = data['strike'].apply(lambda x: x if x.isdigit() else None)


data

nifty = data[data['symbol'].str.contains(r'\bNIFTY', case=False)]


nifty_options = nifty[nifty['options_type']!='FUT']

nifty_options

nifty_options = nifty_options.sort_values(by=['symbol', 'date'])

```

```python
nifty_options
```

```python
import re
expiry_pattern = r'(\d{2}[A-Z]{3}\d{2})'
nifty_options['expiry_date'] = nifty_options.apply(lambda row: re.search(expiry_pattern, row['symbol']).group(0) if row['options_type'] in ['PE', 'CE'] else None, axis=1)
```

```python
nifty_options['expiry_date'] = pd.to_datetime(nifty_options['expiry_date'], format='%d%b%y')
```

```python
nifty_options.to_csv('nifty_options_eg.csv')
```
